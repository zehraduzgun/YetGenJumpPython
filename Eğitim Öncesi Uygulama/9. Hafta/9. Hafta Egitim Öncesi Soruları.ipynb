# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, color, make, model):
        self.color = color
        self.make = make
        self.model = model

class Bus(Vehicle):
    def __init__(self, color, make, model):
        super().__init__(color, make, model)
        self.seating_capacity = 50
        
    def seating_capacity(self):
        print(f"This bus has a seating capacity of {self.seating_capacity} passengers.")
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self, color, make, model):
        self.color = color
        self.make = make
        self.model = model

class Bus(Vehicle):
    def __init__(self, color, make, model):
        super().__init__(color, make, model)
        self.seating_capacity = 50
        
    def seating_capacity(self):
        print(f"This bus has a seating capacity of {self.seating_capacity} passengers.")

School_bus = Bus("yellow", "Volvo", "B7L")
print(isinstance(School_bus, Vehicle))  # True
````
